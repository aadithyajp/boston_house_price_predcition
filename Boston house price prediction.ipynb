```python
# independent faeture and depending feature
X = boston.iloc[:,:-1]
y = boston.iloc[:,-1]
bostonx = boston.iloc[:,:-1]
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

# model training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
# print the coefficient and the intercept 
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which parameter the model has been trained
regression.get_params()
```

```python
y_predict = regression.predict(X_test)
```

```python
y_predict
```

```python
#plot for prediction
plt.scatter(y_test,y_predict)
```

```python
# residual 
residuals = y_test - y_predict
```

```python
residuals
```

```python
# plot of residuals
sns.displot(residuals,kind="kde")
```

```python
# scatter plot with respect to prediction and residuals
#uniform distribution 
plt.scatter(y_predict,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(np.sqrt(mean_squared_error(y_test,y_predict)))
```

# Assumption

```python
# Rsquare and adjusted R square
from sklearn.metrics import r2_score
score = r2_score(y_test,y_predict)
score
```

```python
#adjusted R square = 1-[(1-r2)*(n-1)/(n-k-1)]
#r2:the r2 of the model , n:the number of observation, k:thte number of predictor variables 
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
